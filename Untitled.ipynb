```python
#! /usr/bin/python3
# weather.py - Scrape weather history from www.wunderground.com


import requests
import bs4
import collections
import pandas as pd




#    airport = input('What airport code (ICAO) do you want the weather for? ')
#    start_date = input('The beginning date (2017/01/15) of the weather history: ')
#    end_date = input('The end date (2017/05/21) of the weather history (1 year max): ')

    
    

def header():
    print()
    print('-------------------------------')
    print('        Weather App')
    print('-------------------------------')
    print()


def get_html_from_web(airport, start_date, end_date):
    url = 'https://www.wunderground.com/history/airport/{}/{}/CustomHistory.html?dayend={}&monthend={}&yearend={}'.format(
        airport, start_date, end_date[8:], end_date[5:7], end_date[0:4]
    )
    response = requests.get(url)

    return response.text


def get_data_from_html(html):
    soup = bs4.BeautifulSoup(html, 'html.parser')
    table = soup.find(id='obsTable')
    df = pd.read_html(str(table), skiprows=0, header=1)

    return table


header()
airport = 'SBGR'
start_date = '2015/01/01'
end_date = '2015/12/31'
html = get_html_from_web(airport, start_date, end_date)
table = get_data_from_html(html)
```

```python
type(table)
```

```python
n_columns = 0
n_rows=0
column_names = []
    
    # Find number of rows and columns
    # we also find the column titles if we can
for row in table.find_all('tr'):
                
    # Determine the number of rows in the table
    a_tags = row.find_all('a')
    td_tags = row.find_all('td')
    if len(a_tags) > 0:
        n_rows+=1
        
        n_columns = len(td_tags)
```

```python
df = pd.DataFrame(columns =range(0,n_columns), index= range(0,n_rows))
row_marker = 0
for row in table.find_all('tr'):
    if row.a:
        column_marker = 0
        columns = row.find_all('td')
        for column in columns:
            df.iloc[row_marker,column_marker] = column.get_text().strip()
            column_marker += 1
        if len(columns) > 0:
            row_marker += 1
```
